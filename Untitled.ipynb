```python
wszystkie_sale = []
```

```python
def wyswietl_sale(sala_do_znalezienia):
    for sala in wszystkie_sale:
        if sala["numer_sali"] == sala_do_znalezienia:
            uklad = sala["uklad"]
            rzad = len(uklad)
            kolumny = len(uklad[0])
            print("",end ="  ")
            for kolumna in range(kolumny):
                print(f"{kolumna+1}",end=" " )
            print()
            for index,rzad in enumerate(uklad):
                print(f"{index+1} {" ".join(rzad)}")
        
```

```python
def stworz_sale(miejsce, rzad, nr_sali):
    for sala in wszystkie_sale:
        if sala["numer_sali"] == nr_sali:
            print(f"Błąd, sala {nr_sali} już istnieje")
            return
    sala_tworzona = [["0" for _ in range(rzad)] for _ in range(miejsce)]
    dane_sali = {
        "numer_sali" : nr_sali,
        "uklad" : sala_tworzona}
    wszystkie_sale.append(dane_sali)
        
    
```

```python
def usun_sale(nr_sali):
    for sala in wszystkie_sale:
        if sala["numer_sali"] == nr_sali:
            wszystkie_sale.remove(sala)
            return
    print("Nie ma takiej sali")
```

```python
def czy_wolne(rzad,miejsce,nr_sali):
    for sala in wszystkie_sale:
        if sala["numer_sali"] == nr_sali:
            r = rzad - 1
            m = miejsce - 1
            if 0 <= r < len(sala["uklad"]) and 0<= m < len(sala["uklad"][0]):
                if sala["uklad"][r][m] == "X":
                    print(f"Rząd {rzad}, miejsce: {miejsce}, w sali nr: {nr_sali} jest zajęte")
                else:
                    print(f"Rząd {rzad}, miejsce: {miejsce}, w sali nr: {nr_sali} jest wolne!")
            else:
                print("Błędne współrzędne")
            return
            
            
```

```python
czy_wolne(1,1,11)
```
